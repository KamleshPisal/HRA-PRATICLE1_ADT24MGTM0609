```python
attrition_by_department = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()
attrition_by_department['Attrition Rate'] = attrition_by_department['Yes'] * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=attrition_by_department.index, y=attrition_by_department['Attrition Rate'], palette='viridis')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.ylim(0, 30) # Set a reasonable y-limit based on the data
plt.show()
```
